```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("data\\puntajes.csv" )
```

Convertir la columna puntaje a tipo de datos numerico



```python
df['Puntaje'] = df['Puntaje'].astype(str).str.strip()
df['Puntaje'] = pd.to_numeric(df['Puntaje'], errors='coerce')
```

Mostrar los 10 puntajes más altos

```python
top_puntajes = df.nlargest(10,'Puntaje')
top_puntajes = top_puntajes.drop_duplicates()
top_puntajes = top_puntajes.drop_duplicates(subset=['Carrera', 'Puntaje'])
top_puntajes
```

Mostrar los 10 puntajes más bajos

```python
top_lower = df.nsmallest(10, "Puntaje")
top_lower = top_lower.drop_duplicates()
top_lower = top_lower.drop_duplicates(subset=['Carrera', 'Puntaje'])
```

Crear un gráfico de barras para los puntajes más altos

```python
plt.figure(figsize=(12,8))
plt.barh(top_puntajes['Carrera'] + ' - ' + top_puntajes['Universidad'],top_puntajes['Puntaje'], color='skyblue')
plt.xlabel('Puntaje')
plt.ylabel('Carrera - Universidad')
plt.title("Top 10 puntajes de corte más altos de ingreso a la Universidad")
plt.gca().invert_yaxis()
plt.show()
```

Gráfico para los puntajes más bajos

```python
plt.figure(figsize=(12,8))
plt.barh(top_lower['Carrera'] + ' - ' + top_lower['Universidad'], top_lower['Puntaje'], color='orange')
plt.xlabel('Puntaje')
plt.ylabel('Carrera - Universidad')
plt.title("Top 10 puntajes de corte más bajos de ingreso a la universidad")
plt.gca().invert_yaxis()
plt.show()
```
